**Adding A Row: Very important to know**

```python
df
```

**ignore_index = True else it won't know the next implicit index to put with the new row. It is not inplace. Ignores the old index, creates a new one.**

```python
#append
dictionary = {'country': 'Zimbabwe',
              'year': 2012,
              'population': 22311143,
              'continent': 'Africa',
              'life_exp': 44.4,
              'gdp_cap': 500.54543534}
df = df.append(dictionary, ignore_index = True) # The new df is assigned back to the old one.
df # Notice how ignore_index = True removed the explicit indexing. 
```

Why did we use df.loc here to create new rows? This is because loc can create new rows without focusing on the sequence of the index. If the last index is 111, we can put 999 at the end and it will work. These are explicit indexes being created by loc and are independent of each other. This is helpful when transaction ID's are to be stored. They must be independent of each other.

**Handling Duplicates:**

```python
df.duplicated().tail(10) # It tells you the duplicates except the first occurence is False.
```

```python
df.loc[df.duplicated()] # We're masking with bool values to return the duplicates.
```

```python
df.drop_duplicates(keep = "first") # Keeps the first occurence. row with indexes 1705, 1720, 1723 gone!
```

```python
df.drop_duplicates(keep = "last") # 1705 and 1723 not dropped.
```

```python
df.drop_duplicates(keep = False) # Drops all occurences
```

```python
df
```

```python
df.drop_duplicates(subset = ['country'], keep = 'last') # drop duplicates based on 'country'.
```
