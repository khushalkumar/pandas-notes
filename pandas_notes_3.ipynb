**Concatenating:**

```python
x = pd.DataFrame({'A': [10, 30], 'B': [20, 40]}) # Creating a df
y = pd.DataFrame({'A': [10, 30, 100], 'C': [20, 40, 100], 'D': [200, 200, 300]})
x
```

```python
y
```

axis = 0, vertical direction. Imagine boxes on top of one another. 

Comparison of columns is done here.

```python
df4 = pd.concat([x,y]) # concatenating with axis = 0 as default.
df4 # Notice the indexing is not uniform. It is explicit. Let's make it implicit.
```

```python
df4 = pd.concat([x,y], ignore_index = True) # making index implicit.
df4
```

```python
df4.info() # columns B C and D are float because NaN is float type. 
```

axis = 1, horizontal direction.

Comparison of indexes is done here.

```python
pd.concat([x,y], axis = 1)
```

**Merge:**

concat: stack multiple df along 0/1 axis.

merging: combines df in smarter way based on values of columns.

image.png

```python
df
```

```python
users = pd.DataFrame({'userid': [1,2,3], 'name': ["sarath","shahid","khushal"]})
```

```python
users
```

```python
msgs = pd.DataFrame({'userid': [1,1,2,4], 'msg': ['hmm', 'accha', 'theek hai', 'nice']})
```

```python
msgs
```

```python
users.merge(msgs, on = 'userid') # default 'how' for merge is inner.
```

```python
users.merge(msgs, on = "userid", how = "outer") # outer merge.
```

```python
users.merge(msgs, on = "userid", how = "left")
```

```python
users.merge(msgs, on = 'userid' ,how = "right")
```

if the two dataframes don't have a similar named column:

```python
users.rename(columns = {'userid': 'id'}, inplace = True) # changing userid label to id.
```

```python
msgs
```

```python
users
```

```python
users.merge(msgs, left_on = "id", right_on = 'userid') # two columns with similar elements appear.
# Later on, we can drop one of these columns.
```

First, let's check if all director names in 'movies' are also available in 'directors':

```python
import numpy as np
```

```python
movies['director_id'].isin(directors['id']) # These are individual values. We want one final bool val.
```

```python
np.all(movies['director_id'].isin(directors['id'])) # Checks all values and gives the output.
```

Now, we can merge:

```python
data = movies.merge(directors, left_on = 'director_id', right_on = 'id', how = 'inner') # left will work the same.
```

How conflicting column names are handled:

```python
data # both dataframes had a column labeled 'id'. Hence, the id from left df while merging becomes id_x and from right df, id_y.
```

```python
data.drop(columns = ['director_id', 'id_y'], inplace = True ) # dropping unwanted id columns.
```
