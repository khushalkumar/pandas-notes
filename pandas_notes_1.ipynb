head(): This function returns the first `n` rows for the object based
on position. It is useful for quickly testing if your object
has the right type of data in it.

```python
df.head() # Return the first `n` rows. Default n is 5.
```

```python
df.head? # Read documentation 💎
```

tail(): This function returns last `n` rows from the object based on
position. It is useful for quickly verifying data, for example,
after sorting or appending rows.

```python
df.tail() # 
```

```python
df.tail? # READ THE DOCUMENTATION! 
```

```python
df.head(12) # Return the first 12 rows
```

```python
df.tail(12) # Return the last 12 rows
```

For negative values of `n`, this function returns all rows except
the first `n` rows, equivalent to ``df[n:]``.

```python
df.tail(-10) # all rows except the first 10 rows.
```

For negative values of `n`, head() function returns all rows except
the last `n` rows, equivalent to ``df[:-n]``.

```python
df.head(-10) # all rows except the last 10 rows
```

**This is how I remember negative values in head/tail (Ignore this if not interested):** 

For -ve values, `head` will always give values (that have head in it) starting from head(0 index), which means last 10 will be excluded in case of `head(-10)`.

For -ve values, `tail` will always give values that have tail(last rows) in it, which means first 10 will be excluded in case of `tail(-10)`.

In other words, `a data frame is a collection of series having the same index.`

**Indexing dilemma:** 

So, 0th index does not exist. This is because when we access a particular element, say `ser[0], explicit indexing is being used`, which we defined ie. range(1,1705). 

But when we use slicing, say `ser[0:3] it uses implicit indexing`, ie. range(0,1704) in this case.  

Confused? 😵😵 You won't be after a few cells. 

```python
ser[0:3] # It is using implicit indexing ie. 0 based indexing. Elements will be from 0th to 2nd index.
```

```python
ser[1:4] # it'll give elements from index 1 to index 3, implicit indexing being used. 
```

# loc( explicit) , iloc(implicit)

Now's the time. All the indexing dilemma will be resolved now. `loc and iloc` is your answer to the inconsistent behavior of Pandas when accessing elements.

```python
ser2 # Accessing/printing 'ser2' Series.
```

```python
ser2.loc[5] # loc will always give you explicitly (by us) defined indexes. 
```

```python
ser2.loc[4:6] # Notice that 6 is inclusive. loc will behave the same with slicing as well, giving you what you asked for using explicit indexing.
```

```python
ser3 # let's play with string/character indexes.
```

```python
ser3['f'] # We know this takes explicit index. 
```

```python
ser3.loc['f'] # Takes explicit index
```

```python
ser3.loc['c':'g'] # Takes explicit index
```

```python
ser2
```

implicit vs explicit Indexing.jpg

**How to remember what iloc does?** 

`i - implicit`

Thus, iloc will take implicit indexes only.

```python
ser2.iloc[3] # Check the implicit index 3 in the table above.
```

```python
ser2.iloc[1:4] # See the consistency. iloc will take implicit index slicing from 1 to 3, 4 is excluded.
```

**Slicing rule: ser[1:4]**

loc - last val(4) in inclusive

iloc - last val(4) is exclusive.(not considered)

```python
ser3
```

implicit vs explicit Indexing – string indexing.jpg

```python
ser3.iloc[4] # Check the table above for clear understanding. Remember i - means implicit.
```

```python
ser3.iloc[2:7] # Still consistent.
```

```python
ser3.iloc["d":"h"] # Since iloc takes implicit indexes, check the table above, do you see characters in the implicit indexes? No. Thus, it won't work.
```

```python
ser3.loc['d':'h'] # This would work since loc takes explicit indexes.
```
