Find out all the movies of those directors who have created atleast one movie budgeted > 100 million dollars.

image.png

**Explanation:**

x = 'director_name'

max budgeted movie of each director is compared with >= 100.

filter will return T or F.

If returned T, all movies of that director will be included.

If F, all movies of that director will be filtered out.


```python
# filter is working on the entire sub-group, not a single value.
df_high_budget_dir = data.groupby('director_name').filter(lambda x:x['budget'].max() >= 100) 
```

```python
df_high_budget_dir
```

Another way to find movies of high budgeted directors would be:

```python
high_budget_movies = data.loc[data['budget'] >= 100]
high_budget_movies
```

```python
unique_high_budget_dir = high_budget_movies['director_name'].unique()
unique_high_budget_dir
```

```python
data.loc[data['director_name'].isin(unique_high_budget_dir)]
```

**size in pandas:**

```python
import pandas as pd
student_data = pd.DataFrame({
'school_code': ['s001','s002','s003','s001','s002','s004','s001','s003'],
'class': ['V', 'V', 'VI', 'VI', 'V', 'VI','IX','X'],
'name': ['Alberto Franco','Gino Mcneill','Ryan Parkes', 'Eesha Hinton', 'Gino Mcneill', 'David Parkes','Jane','Martin'],
'age': [12, 12, 13, 13, 14, 12, 15, 20],
'height': [173, 192, 186, 167, 151, 159, 140, 160],
'weight': [35, 32, 33, 30, 31, 32, 40, 50]},
 index=['S1', 'S2', 'S3', 'S4', 'S5', 'S6','S7','S8']) 
```

```python
student_data
```

```python
student_data.size # r * c
```

```python
grouped_data = student_data.groupby(['school_code', 'class'])      
print(grouped_data.size())  # number of rows for each group.
print(grouped_data.count()) # number of values. Here it is based on the groupby.
```

**groupby.tail: Return last n rows of each group.**

```python
student_data
```

```python
n = 1                                                              
rows_to_drop=student_data.groupby(['school_code']).tail(n).index
print(rows_to_drop)

indexes_to_be_deleted = student_data.groupby(['school_code']).tail(n)
print(indexes_to_be_deleted) 

result_of_drop = student_data.drop(rows_to_drop, axis=0)
print(result_of_drop)
```
